```python
import folium
import pandas as pd
import numpy as np
import geopandas as gpd

# make a geojason df for sd
geo_data_file = r'/data/bp-codeathon-team-2/sd.geojson'
with open(geo_data_file, 'r') as jsonFile:
    geo_df=gpd.read_file(jsonFile)
geo_df = geo_df.rename(columns = {"zip":"zipcode"})
```

```python
geo_df
```

```python
# read the stats file, merge data for these zipcodes, and then return to 
income_data = pd.read_csv("/data/bp-codeathon-team-2/zip_mean_median_income.csv")
income_data
```

```python
# going to transform the income data
income_data["zipcode"] = [item.split("US")[-1] for item in income_data["id"]]
income_data
```

```python
income_data.columns
```

```python
new_income_data = income_data.rename(columns={
                           "Estimate!!Households!!Total": "total_households",
                           "Estimate!!Households!!Median income (dollars)": "median_household_income",
                           "Estimate!!Households!!Mean income (dollars)": "mean_household_income",
                           "Estimate!!Households!!PERCENT ALLOCATED!!Household income in the past 12 months": "income_percent_allocated"})[["zipcode", "total_households", "median_household_income", "mean_household_income", "income_percent_allocated"]]
```

```python
new_income_data
```

```python
# read the ZCTA.geojson from cdc

# make a geojason df for sd
cdc_data_file = r'/data/bp-codeathon-team-2/ZCTA.geojson'
with open(cdc_data_file, 'r') as jsonFile:
    cdc_df=gpd.read_file(jsonFile)
cdc_df = cdc_df.rename(columns = {"zcta5":"zipcode"}).drop(columns=["geometry"])
```

```python
cdc_df.columns
```

```python
merge_1 = geo_df.merge(new_income_data, on="zipcode", how="left")
merge_1
```

```python
merge_2 = merge_1.merge(cdc_df, on="zipcode", how="left")
merge_2
```

```python
merge_2[merge_2["total_households"].isna()]
```

```python
# make a geojason from merged_data
cdc_data_file = r'/data/bp-codeathon-team-2/sd_zip_complete.geojson'
merge_2.to_file(cdc_data_file, driver="GeoJSON")
```
