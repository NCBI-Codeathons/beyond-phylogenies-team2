# Andersen Lab Genomic Data Filtering

Using publicly available genomic data from the Andersen lab @ scripps (available here: https://github.com/andersen-lab/HCoV-19-Genomics)

Filtering down to just San Diego based sequeneces that have a fasta file name

```python
# imports
import pandas as pd
import glob
import os
```

```python
# define file locations
consensus_sequences_path = "../data/consensus_sequences/"
metadata_path = "../data/sd_sequence_metadata.csv"
```

```python
# load the metadata file and get the sequence names for file filtering
metadata = pd.read_csv(metadata_path)
metadata
```

```python
# get the file names from the 'fasta_hdr' column

def fasta_hdr_to_filename(fasta_hdr_list):
    data_list = []
    for item in fasta_hdr_list:
        try:
            data_list.append(item.split("/")[2] + ".fasta")
        except:
            print(f"failed - {item}")
    return data_list

sequence_file_names = fasta_hdr_to_filename(metadata['fasta_hdr'])
```

```python
len(sequence_file_names)
```

```python
list_all_consensus_file = glob.glob(consensus_sequences_path + "*.fasta")
len(list_all_consensus_file)
```

```python
for file in list_all_consensus_file:
    if os.path.basename(file) not in sequence_file_names:
        os.remove(file)
        print(f"removing non-sd file {file}")
```

```python
remaining_files = [os.path.basename(file) for file in glob.glob(consensus_sequences_path + "*.fasta")]
len(remaining_files)
```

```python
missing_files = []
for file in sequence_file_names:
    if file not in remaining_files:
        missing_files.append(file)
missing_files
```
