```python
import folium
import pandas as pd
import numpy as np
import geopandas as gpd

sd_zip_file = r'/data/bp-codeathon-team-2/sd_zip_complete.geojson'
with open(sd_zip_file, 'r') as jsonFile:
    sd_zip_df=gpd.read_file(jsonFile)
    
sd_zip_df = sd_zip_df.dropna()

sd_zip_df_income = sd_zip_df[sd_zip_df['median_household_income'].str.isnumeric()]
sd_zip_df_income['median_household_income'] = sd_zip_df_income['median_household_income'].astype(int)

# make a geojason df for sd
geo_data_file = r'/data/bp-codeathon-team-2/sd.geojson'
with open(geo_data_file, 'r') as jsonFile:
    geo_df=gpd.read_file(jsonFile)
geo_df = geo_df.rename(columns = {"zip":"zipcode"})

# create the initial map (total 6 sub-maps)
sd_map_income = folium.Map(location=[32.7157, -117.1611], zoom_start=8)
folium.TileLayer('CartoDB positron',name="Light Map",control=False).add_to(sd_map_income)

sd_map_obe = folium.Map(location=[32.7157, -117.1611], zoom_start=8)
folium.TileLayer('CartoDB positron',name="Light Map",control=False).add_to(sd_map_obe)

sd_map_dia = folium.Map(location=[32.7157, -117.1611], zoom_start=8)
folium.TileLayer('CartoDB positron',name="Light Map",control=False).add_to(sd_map_dia)

sd_map_2_12 = folium.Map(location=[32.7157, -117.1611], zoom_start=8)
folium.TileLayer('CartoDB positron',name="Light Map",control=False).add_to(sd_map_2_12)

sd_map_2 = folium.Map(location=[32.7157, -117.1611], zoom_start=8)
folium.TileLayer('CartoDB positron',name="Light Map",control=False).add_to(sd_map_2)

sd_map_5 = folium.Map(location=[32.7157, -117.1611], zoom_start=8)
folium.TileLayer('CartoDB positron',name="Light Map",control=False).add_to(sd_map_5)


## create the choropleth map
choropleth1 = folium.Choropleth(
        geo_data=geo_df,
        data=sd_zip_df_income,
        columns=['zipcode','median_household_income'],
        key_on='feature.properties.zipcode',
        nan_fill_color="White",
        fill_color='YlGn',
        fill_opacity=0.7,
        line_opacity=0.8,
        line_color='Blue',
        legend_name='number of cases',
        highlight=True
    ).add_to(sd_map_income)


sd_zip_df_obe = sd_zip_df.copy()
sd_zip_df_obe['obesity_crudeprev'] = sd_zip_df_obe['obesity_crudeprev'].astype(float)

choropleth_obe = folium.Choropleth(
        geo_data=geo_df,
        data=sd_zip_df_obe,
        columns=['zipcode','obesity_crudeprev'],
        key_on='feature.properties.zipcode',
        nan_fill_color="White",
        fill_color='YlGn',
        fill_opacity=0.7,
        line_opacity=0.8,
        line_color='Blue',
        legend_name='number of cases',
        highlight=True
    ).add_to(sd_map_obe)


sd_zip_df_dia = sd_zip_df.copy()
sd_zip_df_dia['diabetes_crudeprev'] = sd_zip_df_dia['diabetes_crudeprev'].astype(float)

choropleth_dia = folium.Choropleth(
        geo_data=geo_df,
        data=sd_zip_df_dia,
        columns=['zipcode','diabetes_crudeprev'],
        key_on='feature.properties.zipcode',
        nan_fill_color="White",
        fill_color='YlGn',
        fill_opacity=0.7,
        line_opacity=0.8,
        line_color='Blue',
        legend_name='number of cases',
        highlight=True
    ).add_to(sd_map_dia)



# the first_appearance of variants data
appear_df = pd.read_csv('first_appearance.csv')
df_2_12 = appear_df[appear_df['nextclade_pango'] == 'BA.2.12.1'][['community','week_appeared']]

choropleth_2_12 = folium.Choropleth(
        geo_data=geo_df,
        data=df_2_12,
        columns=['community','week_appeared'],
        key_on='feature.properties.community',
        nan_fill_color="White",
        fill_color='YlGn',
        fill_opacity=0.7,
        line_opacity=0.8,
        line_color='Blue',
        legend_name='number of cases',
        highlight=True
    ).add_to(sd_map_2_12)


df_2 = appear_df[appear_df['nextclade_pango'] == 'BA.2'][['community','week_appeared']]

choropleth_2 = folium.Choropleth(
        geo_data=geo_df,
        data=df_2,
        columns=['community','week_appeared'],
        key_on='feature.properties.community',
        nan_fill_color="White",
        fill_color='YlGn',
        fill_opacity=0.7,
        line_opacity=0.8,
        line_color='Blue',
        legend_name='number of cases',
        highlight=True
    ).add_to(sd_map_2)

df_5 = appear_df[appear_df['nextclade_pango'] == 'BA.5'][['community','week_appeared']]

choropleth_5 = folium.Choropleth(
        geo_data=geo_df,
        data=df_5,
        columns=['community','week_appeared'],
        key_on='feature.properties.community',
        nan_fill_color="White",
        fill_color='YlGn',
        fill_opacity=0.7,
        line_opacity=0.8,
        line_color='Blue',
        legend_name='number of cases',
        highlight=True
    ).add_to(sd_map_5)
```
